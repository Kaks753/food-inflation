## Key Findings

### 1. Price Trends (2006-2021)
- Maize prices increased ~45% (CAGR 2.5%)
- Minimum: 6 KES/kg (2006)
- Maximum: 100 KES/kg (2017 drought)
- Average: 48 KES/kg

### 2. Volatility
- Coefficient of variation: ~35%
- Highest spikes during droughts (2011, 2017)
- Imported commodities (rice, oil) more stable

### 3. Seasonality
- Clear seasonal patterns
- Harvest months (Oct-Dec): 15-20% lower prices
- Lean season (Feb-May): 15-20% higher prices
- Peak-to-trough swing: 15-20 KES/kg

### 4. Regional Variation
- Nairobi & Coast: Higher prices (50-55 KES/kg avg)
- Rift Valley: Lower prices (40-45 KES/kg avg)
- Price gap: 20-30 KES/kg between regions

### 5. Model Performance
- Prophet: MAPE ~8-12%
- SARIMA: MAPE ~7-10%
- XGBoost: MAPE ~6-9%
- Ensemble approach recommended

### 6. 12-Month Forecast
- Expected price range: 45-55 KES/kg
- Anticipated inflation: +3-7%
- Seasonal patterns likely to persist

## Recommendations

### For Policymakers
1. **Strategic Grain Reserves**: Build 3-6 month buffer stocks
2. **Market Integration**: Improve transport infrastructure (Rift Valley → Nairobi/Coast)
3. **Early Warning System**: Implement price monitoring dashboard
4. **Subsidy Timing**: Target lean season (Feb-May) for maximum impact
5. **Regional Balance**: Facilitate inter-regional trade to reduce price gaps

### For Farmers
1. **Harvest Timing**: Consider storage facilities to sell during lean season (+15-20% premium)
2. **Crop Diversification**: Reduce exposure to maize price volatility
3. **Market Information**: Use price forecasts for planting decisions
4. **Cooperatives**: Form buying/selling cooperatives to reduce transaction costs

### For Consumers
1. **Bulk Buying**: Purchase during harvest months (Oct-Dec) for 15-20% savings
2. **Storage**: Invest in proper storage for 3-6 month supply
3. **Alternatives**: Consider substitute staples when maize prices spike
4. **Budget Planning**: Expect +3-7% annual maize price inflation

### For Traders & Businesses
1. **Arbitrage**: Exploit regional price differences (20-30 KES/kg gap)
2. **Inventory Management**: Stock up during harvest, sell during lean season
3. **Hedging**: Use price forecasts to hedge commodity risk
4. **Supply Chain**: Optimize logistics to reduce regional price gaps
